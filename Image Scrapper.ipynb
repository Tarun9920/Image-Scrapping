```python
def make_soup(url):
    page= requests.get(url)
    soup= bs(page.content,'html.parser')
    return soup
```

```python
# scrap images per link
images= []

for i in range(len(url)):
    soup= make_soup(url[i])
    for p in soup.findAll('div',class_="c6e09d"):
        for j in p.findAll('meta',itemprop='image'):
            content= j.get('content')
            if content is not None and "jpg" in content:
                content= content[:-6]
                content= content+"480.jpg"
                images.append(content)
                print content
```
